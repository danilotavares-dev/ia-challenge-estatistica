```python
# importar as bibliotecas necessarias
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
```

```python
df_faturamento
```

```python
df_faturamento.info()
```

```python
df_faturamento['data_ref'] = pd.to_datetime(df_faturamento['data_ref'])
```

```python
df_faturamento.info()
```

### Média - Vendas

```python
# mádia de vendas
media_vendas = round(df_faturamento['valor'].mean(), 2)
media_vendas
```

### Gráfico de barra vertical - Vendas

```python
df_faturamento.sort_values(by='data_ref')
```

```python
df_faturamento['mes_ano'] = df_faturamento['data_ref'].dt.strftime('%b-%Y')
```

```python
df_faturamento.plot.bar(x='mes_ano', y='valor', figsize=(12, 7), legend='False', title='Faturamento Mensal')
```

### Gráfico de linhas - Vendas

```python
df_faturamento.plot.line(x='mes_ano', y='valor', figsize=(12, 7), legend='False', title='Faturamento Mensal', )
```
